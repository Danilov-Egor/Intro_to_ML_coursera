```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
```

```python
prices = pd.read_csv("close_prices.csv", index_col="date")
```

```python
prices
```

```python
dj = pd.read_csv("djia_index.csv", index_col="date")
```

```python
dj.head()
```

```python
for n in range(1,10):
    pca = PCA(n_components=n)
    pca.fit(prices)
    print(n, ":", (pca.explained_variance_ratio_).sum())
```

```python
pca = PCA(n_components=10)
pca.fit(prices)
prices_transformed = pd.DataFrame(pca.transform(prices))
```

```python
prices_transformed
```

```python
(pca.explained_variance_ratio_)
```

```python
first_comp = prices_transformed[0]
dj_index = dj['^DJI']
```

```python
np.corrcoef(first_comp, dj_index)
```
